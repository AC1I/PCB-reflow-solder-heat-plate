```python
import numpy as np
import matplotlib.pyplot as plt
```

### LMT85 study
A couple goals here. Determine if the linearization of the LMT85 response curve is sufficient

```python
voltages = np.linspace(1365, 310, num=1000)

def linearfunc(v):
    return (v - 1365.) / ((301. - 1365.)/(150. - 25.)) + 25.

def nonlinearfunc(v):
    squared_section = np.sqrt(8.194**2.0 + 4*0.00262*(1324 - v))
    return (8.194 - squared_section)/(2*-0.00262) + 30

x_linear = linearfunc(voltages)
x_nonlinear = nonlinearfunc(voltages)
error = x_nonlinear - x_linear

plt.figure()
plt.plot(voltages, x_linear, label="linear")
plt.plot(voltages, x_nonlinear, label="nonlinear")
plt.legend()
plt.title("theoretical linear vs nonlinear graph")
plt.show()

plt.figure()
plt.plot(voltages, error, label="error")
plt.title("error between linear and nonlinear transfer functions")
plt.show()
```

# Conclusion
The conclusion is that the linear function is fine for this application. It has a maximum error of ~1.3C over the range
